```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('Hotel_recomender_system_based_on_zomato_data.csv')
df
```

```python
df.drop(columns=['menu_item', 'listed_in(type)', 'listed_in(city)'], inplace=True)
```

```python
df = df.dropna()
```

```python
df
```

```python
df.columns
```

```python
def place(city):
    df.loc[:, 'location'] = df['location'].str.lower()
    place=df[df['location']==city.lower()]
    place=place.sort_values(by='rate',ascending=False)
    if(place.empty==0):
        hname=place[['address','name','rate','votes', 'phone' ,'location','rest_type', 'url']]
        return hname.head()
    else:
        print('No Hotels Available')
```

```python
print('find your hotel place')
place("Basavanagudi")
```
